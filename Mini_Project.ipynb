```python
stock,basket={},{}

def add(ProductName, UnitPrice):
  stock[ProductName] = UnitPrice

def update(ProductName, UpdatedPrice):
  stock[ProductName] = UpdatedPrice

def remove(ProductName):
  del stock[ProductName]

def view():
  print(stock)

def addbasket(ProductName , Quantity):
  basket[ProductName] = Quantity

def viewbasket():
  print(basket)

def search(ProductName):
  for i,j in stock:
    if i == ProductName:
      print("Present")
    else:
      print("Oops!!!Sorry not found")

def removebasket(ProductName):
  del basket[ProductName]

def invoice():
  sum = 0
  for i,j in basket:
    print(i," = ",j," * ",stock[i]) 
    sum += j * stock[i]
  print("Total =",sum)

def AdminModule():
  print("Inside Admin")
  ch = 0
  while ch != 5:
    print("Select:\n1-Add Product in Stock (ProductName, UnitPrice)\n2-Update Product price in Stock\n3-Remove Product from Stock\n4-View All Products in Stock\n5-Logout Admin")
    ch = int(input())
    if ch == 1:
      ProductName = input("Product Name:")
      UnitPrice = int(input("Unit Price:"))
      add(ProductName, UnitPrice)
    elif ch == 2:
      ProductName = input("Product Name:")
      UpdatedPrice = int(input("Updated Price:"))
      update(ProductName, UpdatedPrice)
    elif ch == 3:
      ProductName = input("Product Name:")
      remove(ProductName)
    elif ch == 4:
      view()
  print("Logged out of Admin")

def ConsumerModuleBasket():
  print("Inside Consumer Basket")
  ch = 0
  while ch != 7:
    print("Select:\n1-View All Products in Stock\n2-Add to shopping basket\n3-View all product of basket\n4-Search product in Stock\n5-Remove product from basket\n6-Print invoice\n7-Signout")
    ch = int(input())
    if ch == 1:
      view()
    elif ch == 2:
      ProductName = input("Product Name:")
      Quantity = int(input("Quantity:"))
      addbasket(ProductName , Quantity)
    elif ch == 3:
      viewbasket()
    elif ch == 4:
      ProductName = input("Product Name:")
      search(ProductName)
    elif ch == 5:
      ProductName = input("Product Name:")
      removebasket(ProductName)
    elif ch == 6:
      invoice()
  print("Signed out of Consumer")

print("Welcome!!!".center(20),"\nOnline Retail Shopping")
print("Select:\n1.Admin\n2.Consumer")
ch = int(input())
if ch == 1:
  AdminModule()
elif ch == 2:
  ConsumerModuleBasket()
```
